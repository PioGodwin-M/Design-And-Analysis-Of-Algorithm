```python
import matplotlib.pyplot as plt
import networkx as nx
import pylab

# Create an empty undirected weighted graph
G = nx.Graph()

# Add nodes
nodes_list = [1, 2, 3, 4, 5, 6, 7]
G.add_nodes_from(nodes_list)

# Add weighted edges
edges_list = [
    (1, 2, 1), (1, 4, 4), (2, 3, 2), (2, 4, 6), (2, 5, 4),
    (3, 5, 5), (3, 6, 6), (4, 5, 3), (4, 7, 4), (5, 6, 8),
    (5, 7, 7)
]
G.add_weighted_edges_from(edges_list)

# Draw the graph
pos = nx.spring_layout(G)
pylab.figure(1)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12)

# Draw edge labels
nx.draw_networkx_edge_labels(
    G, 
    pos, 
    edge_labels=nx.get_edge_attributes(G, 'weight')
)

# Calculate a minimum spanning tree using Prim's algorithm
mst = nx.minimum_spanning_tree(G, algorithm='prim')
print("Edges in the minimum spanning tree:", sorted(mst.edges(data=True)))

plt.show()
```
